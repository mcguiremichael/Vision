```python
import numpy as np
import scipy
import scipy.ndimage
import cv2
import matplotlib
import matplotlib.pyplot as plt

high_pass_std = 1.0
low_pass_std = 1.0

image1 = cv2.imread('c1.jpg',0)
image2 = cv2.imread('c2.jpg',0)


```

```python
cv2.imshow("First image", image1)
cv2.imshow("Second image", image2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
print(type(image1))
print(image1.dtype)

print(np.max(image1), np.min(image1), np.max(image2), np.min(image2))
```

```python
high_pass_im1 = scipy.ndimage.gaussian_filter(image1.astype(np.float32), high_pass_std, truncate=3.0)
low_pass_im2 = scipy.ndimage.gaussian_laplace(image2.astype(np.float32), low_pass_std, truncate=3.0)

hybrid_img = 0.5 * high_pass_im1 + 0.5 * low_pass_im2
plt.imshow(high_pass_im1, cmap="gray")
plt.figure()
plt.imshow(low_pass_im2, cmap="gray")
plt.figure()
plt.imshow(hybrid_img, cmap="gray")
plt.show()

print(low_pass_im2)
print(low_pass_im2.round().astype(np.uint8))

cv2.imshow("high_pass_im1", high_pass_im1.round().astype(np.uint8)) 
cv2.imshow("low_pass_m2",low_pass_im2.round().astype(np.uint8))
cv2.imshow("Hybrid",hybrid_img.round().astype(np.uint8))
cv2.waitKey(0)
cv2.destroyAllWindows()
```
